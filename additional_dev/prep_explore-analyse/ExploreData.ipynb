```python
import pandas as pd
import numpy as np
```

```python
import plotly.graph_objects as go

import plotly.express as px

import plotly.figure_factory as ff
import plotly.io as pio

from plotly.subplots import make_subplots
```

```python
df=pd.read_csv('data/S50-Daily-Y09-22.csv',index_col='Date/Time',parse_dates=['Date/Time'],dayfirst=True)
df.info()
```

```python
# plot single
#https://plotly.com/python/line-charts/
def plot_line_timeseries(col_name):
 fig=px.line(df[col_name],title=col_name)
 return fig

fig_macd=plot_line_timeseries('MACD10-20')
fig_macd.show()
        
```

```python
#https://plotly.com/python/distplot/
# plot single
def plot_dist_timeseries(col_name):
 fig=px.histogram(df[col_name])
 return fig

fig_macd=plot_dist_timeseries('SIGNAL10-20-15')
fig_macd.show()
        
```

```python
#https://stackoverflow.com/questions/63265707/plotly-how-to-plot-multiple-lines-with-shared-x-axis
cols_plot=['Price','MACD10-20','RSI-MA15-5']
fig = make_subplots(rows=len(cols_plot), cols=1, shared_xaxes=True)

for i, col in enumerate(cols_plot, start=1):
    fig.add_trace(go.Scatter(x=df[col].index, y=df[col].values), row=i, col=1)
fig.update_layout( width=1000,height=1200)
    
fig.show()

    
```
